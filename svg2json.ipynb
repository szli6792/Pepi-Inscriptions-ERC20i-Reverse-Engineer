```python
#import re
import os
import json
from xml.dom import minidom

def parse_svg(svg_content, type):
    doc = minidom.parseString(svg_content)
    rects = doc.getElementsByTagName("rect")
    data = []
    for rect in rects:
        x = int(rect.getAttribute("x"))
        y = int(rect.getAttribute("y"))
        width = int(rect.getAttribute("width"))
        height = int(rect.getAttribute("height"))
        if (type == 0):
            data.append([x, y, width, height])
        else:
            color = rect.getAttribute("fill")
            process_color = int(color[1:],16)
            data.append([x, y, width, height, process_color])
    return data

def process_folder(folder_path, setLength, j, type):
    json_data = []  # Initialize json_data as a list
    counter = 1
    for filename in os.listdir(folder_path):
        if (setLength*j == 1):
            if filename.endswith(".svg"):
                #match = re.search(r'\((.*?)\)', filename)
                #hex_data = match.group(1)
                body_data = []
                #body_data.append(int(hex_data[0]))
                #body_data.append(int(hex_data[3]))
                #for number in range(set_length
                if (filename[-11] == '_'):
                    body_data.append(int(filename[-9]))
                    body_data.append(int(filename[-6]))
                    print(int(filename[-9]),int(filename[-6]))
                if (filename[-12] == '_'):
                    body_data.append(int(filename[-10]))
                    body_data.append(int(filename[-6])+int(filename[-7])*10)
                    print(int(filename[-10]),int(filename[-6])+int(filename[-7])*10)
                if (filename[-13] == '_'):
                    body_data.append(int(filename[-11]))
                    body_data.append(int(filename[-6])+int(filename[-7])*10+int(filename[-8])*100)
                    print(int(filename[-11]),int(filename[-6])+int(filename[-7])*10+int(filename[-8])*100)
                with open(os.path.join(folder_path, filename), "r") as f:
                    svg_content = f.read()
                    body_data.append(parse_svg(svg_content,type))
                json_data.append(body_data)
        else:
            #print((j-1)*setLength)
            #print((j)*setLength)
            if (((j-1)*setLength) < counter <= ((j)*setLength)):
                if filename.endswith(".svg"):
                    #match = re.search(r'\((.*?)\)', filename)
                    #hex_data = match.group(1)
                    body_data = []
                    #body_data.append(int(hex_data[0]))
                    #body_data.append(int(hex_data[3]))
                    #for number in range(set_length
                    if (filename[-11] == '_'):
                        body_data.append(int(filename[-9]))
                        body_data.append(int(filename[-6]))
                        print(int(filename[-9]),int(filename[-6]))
                    if (filename[-12] == '_'):
                        body_data.append(int(filename[-10]))
                        body_data.append(int(filename[-6])+int(filename[-7])*10)
                        print(int(filename[-10]),int(filename[-6])+int(filename[-7])*10)
                    if (filename[-13] == '_'):
                        body_data.append(int(filename[-11]))
                        body_data.append(int(filename[-6])+int(filename[-7])*10+int(filename[-8])*100)
                        print(int(filename[-11]),int(filename[-6])+int(filename[-7])*10+int(filename[-8])*100)
                    with open(os.path.join(folder_path, filename), "r") as f:
                        svg_content = f.read()
                        body_data.append(parse_svg(svg_content,type))
                    json_data.append(body_data)
                #counter += 1
        counter += 1
    return json_data

svg_type = [0,1,1,1,1,1]

folder_path1 = "./svgs_from_bytecode/body"
folder_path2 = "./svgs_from_bytecode/eye"
folder_path3 = "./svgs_from_bytecode/mouth"
folder_path4 = "./svgs_from_bytecode/hat"
folder_path5 = "./svgs_from_bytecode/accessory"
folder_path6 = "./svgs_from_bytecode/clothing"
folders = [folder_path1, folder_path2, folder_path3, folder_path4, folder_path5, folder_path6]

output_file1 = "./svgs_to_json/setBody"
output_file2 = "./svgs_to_json/setEyes"
output_file3 = "./svgs_to_json/setMouths"
output_file4 = "./svgs_to_json/setHats"
output_file5 = "./svgs_to_json/setAccessories"
output_file6 = "./svgs_to_json/setClothes"
outputs = [output_file1, output_file2, output_file3, output_file4, output_file5, output_file6]

bodySetPackLength = 1
eyeSetPackLength = 1
mouthSetPackLength = 1
hatSetPackLength = 16
accessorySetPackLength = 17
clothingSetPackLength = 19
packLength = [bodySetPackLength,eyeSetPackLength,mouthSetPackLength,hatSetPackLength,accessorySetPackLength,clothingSetPackLength,]

numberOfBodyPacks = 1
numberOfEyePacks = 1
numberOfMouthPacks = 1
numberOfHatPacks = 12
numberOfAccessoryPacks = 9
numberOfClothesPacks = 18
numberOfSetPacks = [numberOfBodyPacks,numberOfEyePacks,numberOfMouthPacks,numberOfHatPacks,numberOfAccessoryPacks,numberOfClothesPacks]

i=0
for folder in folders:
    pack = numberOfSetPacks[i]
    setLength = packLength[i]
    type = svg_type[i]
    for j in range(1, pack+1):
        #json_data = process_folder(folder)
        if ((j == 1) and (setLength == 1)):
            #print(str(j)+"of"+str(setLength))
            json_data = process_folder(folder,setLength,j,type)
            output_name = outputs[i] + ".json"
            print(output_name)
            with open(output_name, "w") as f:
                json.dump(json_data, f, indent=2)
        else:
            #print(str(j)+"of"+str(setLength))
            json_data = process_folder(folder,setLength,j,type)
            output_name = outputs[i] + str(j) + ".json"
            print(output_name)
            with open(output_name, "w") as f:
                json.dump(json_data, f, indent=2)           
    i+=1
```
